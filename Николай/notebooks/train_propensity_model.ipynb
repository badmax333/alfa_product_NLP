# Обучение модели склонности к продуктам (онбординг)

Синтетический датасет на тех же 5000 клиентах, что классификатор P1–P8.

Артефакты: `data/propensity_synthetic.csv`, `models/propensity_lgbm.pkl`.

```python
from pathlib import Path
import sys

ROOT = Path.cwd().resolve()
if ROOT.name == "notebooks":
    ROOT = ROOT.parent
sys.path.insert(0, str(ROOT / "src"))

import pandas as pd
import matplotlib.pyplot as plt

DATA_PATH = ROOT / "data" / "propensity_synthetic.csv"
MODEL_PATH = ROOT / "models" / "propensity_lgbm.pkl"
DATA_PATH
```

```python
if not DATA_PATH.exists():
    from generate_propensity_synthetic import main as gen_main
    gen_main()

df = pd.read_csv(DATA_PATH)
print(df.shape)
df.head(3)
```

```python
rate = df.groupby("product_id")["propensity_label"].mean().sort_values(ascending=False)
rate.plot(kind="bar", title="Доля положительной склонности по продуктам", figsize=(10, 4))
plt.ylabel("P(label=1)")
plt.tight_layout()
```
